```python
import pandas as pd
import numpy as np
import plotly.express as px

sales = pd.read_csv("/Users/jordan.arlan/Documents/DS100/Data/vgsales.csv")
sales
```

```python
sales[["Year", "NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Global_Sales"]].groupby("Year").agg(sum)
```

```python
sales_list = ["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Global_Sales"]
sales[["Genre", "NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Global_Sales"]].groupby("Genre").agg(sum)
```

```python
sales.sort_values("Global_Sales", ascending=False).groupby("Platform").agg(lambda x : x.iloc[0]).sort_values("Year")
```

```python
sales.groupby("Year").size()
```

```python
sales.groupby("Year").filter(lambda sf: sf["Global_Sales"].max() < 5)
```

```python
# Find the total global sales each year for each platform
sales.groupby(["Year", "Platform"])[sales_list].agg(sum)
```

```python
# The `pivot_table` method is used to generate a Pandas pivot table
import numpy as np
sales.pivot_table(index = "Year", columns = "Platform", values = "Global_Sales", aggfunc = np.sum)
```
