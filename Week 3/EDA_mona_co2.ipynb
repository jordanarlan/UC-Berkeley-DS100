```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

co2_file = "Week 3/co2_mm_mlo.csv"
```

```python
co2 = pd.read_csv(co2_file,skiprows=56)
```

```python
co2.iloc[0:2]
```

```python
sns.lineplot(x='decimal date', y='average', data=co2)
```

```python
co2.shape
```

```python
co2["month"].value_counts().sort_index()
```

```python
sns.displot(co2['ndays']);
plt.title("Distribution of days feature")
plt.show() # suppresses unneeded plotting output
```

```python
sns.scatterplot(x="year", y="ndays", data=co2)
plt.title("Day field by Year"); # the ; suppresses output
```

```python
# Histograms of average CO2 measurements
sns.displot(co2['average']);
```

```python
sns.lineplot(x='year', y='average', data=co2.groupby('year').mean())
plt.title("CO2 Average By Year")
```
